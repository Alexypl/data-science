```
import pandas as pd 
import numpy as np
```

```
cov = pd.read_excel("COVID-19.xlsx")
cov
```

```
cov.columns = [col.lower().replace(" ", "_") for col in cov.columns]
cov
```

```
import seaborn as sns
sns.set()

cov = cov.dropna()

cov.nowe_zakażenia.plot()
```

```
cov['wsz_zak_przes'] = cov.wszystkie_zakażenia.shift()
cov.head()
```

```
cov["dzienny_przyrost"] = round((cov.wszystkie_zakażenia / cov.wsz_zak_przes - 1) *100,2)
cov 
```
