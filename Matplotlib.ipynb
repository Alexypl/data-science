```
import matplotlib 
matplotlib.__version__
```

```
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
#sns.set()
```

```
fig = plt.figure()
fig.show()
```

```
_ = plt.plot([0, 1], [0, 3])
```

```
_ = plt.plot([0, 1], [0, 3], label = 'line', color='green')
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.title("Wykres")
plt.legend()
```

```
_ = plt.plot([0, 1, 2], [0, 3, 2], label = 'line', color='green')
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.title("Wykres")
plt.legend()
```

```
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

Tworzenie wielu wykresów


```
plt.subplot(121) # 1 liczba wierszy w siatce wykresów 2 - liczba kolumn 1 - konkretny wykres liczba wykresów

plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 4, 2], 'r--')


```

```
fig = plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 4, 2], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 4, 2], 'ro')
```

Wykres słupkowy

```
plt.bar(x = [0, 1, 2, 3], height=[1.2 , 3.2, 1.4, 0.9])
```

```
plt.bar(x = ["a", "b", "c", "d"], height=[1.2 , 3.2, 1.4, 0.9])
```

```
plt.barh(y = ["a", "b", "c", "d"], width=[1.2 , 3.2, 1.4, 0.9])
```
